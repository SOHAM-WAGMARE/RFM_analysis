```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\hp\Desktop\RFM_analysis\cleaned_data.csv")
df
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

snapshot=df['InvoiceDate'].max() + pd.Timedelta(days=1)
print(snapshot)
```

```python
rfm=df.groupby('CustomerID').agg({
    'InvoiceDate':lambda x:(snapshot - x.max()).days,
    'InvoiceNo':'nunique',
    'TotalPrice':'sum'}).reset_index()
rfm.rename(columns={
    'InvoiceDate':'Recency',
    'InvoiceNo':'Frequency',
    'TotalPrice': 'Monetary'},inplace=True)

rfm.head()
 
```

```python
rfm.to_csv('rfm_table.csv',index=False)
```
